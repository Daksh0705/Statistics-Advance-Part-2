```python
#Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the distribution (like a fair 6-sided die)
low, high = 1, 7  # randint is inclusive of low, exclusive of high
x = np.arange(low, high)

# Compute PMF
pmf_values = randint.pmf(x, low, high)

# Plot the PMF
plt.bar(x, pmf_values, width=0.6, color='skyblue', edgecolor='black')
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks(x)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
from scipy.stats import bernoulli

def bernoulli_pdf(x, p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution.

    Parameters:
    x : int or list of ints
        The value(s) to calculate the PDF for, should be 0 or 1.
    p : float
        Probability of success (1), between 0 and 1.

    Returns:
    float or array
        Probability of x under Bernoulli(p).
    """
    return bernoulli.pmf(x, p)

# Example usage
p = 0.7
print("P(X=0):", bernoulli_pdf(0, p))
print("P(X=1):", bernoulli_pdf(1, p))
```

```python
#Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the discrete uniform distribution range (1 to 6 inclusive)
low, high = 1, 7  # randint is inclusive low, exclusive high

x = np.arange(low, high)  # values 1, 2, ..., 6
cdf_values = randint.cdf(x, low, high)

# Plot the CDF using a step plot to emphasize discrete nature
plt.step(x, cdf_values, where='post', color='purple')
plt.title("CDF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.xticks(x)
plt.ylim(0, 1.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def compute_and_plot_zscores(data):
    """
    Calculate Z-scores and plot them.

    Parameters:
    data : list or array-like
        Input dataset.
    """
    z_scores = zscore(data)

    # Plot Z-scores
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='gray', linestyle='--')
    plt.axhline(1, color='red', linestyle='--', alpha=0.7)
    plt.axhline(-1, color='red', linestyle='--', alpha=0.7)
    plt.title("Z-Scores of Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-Score")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(50, 10, 100)
z = compute_and_plot_zscores(data)
```

```python
#Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    """
    Calculate and plot the PDF of the standard normal distribution (mean = 0, std = 1).
    """
    # Generate x values over a range
    x = np.linspace(-4, 4, 500)

    # Calculate the standard normal PDF
    pdf = norm.pdf(x, loc=0, scale=1)

    # Plot the PDF
    plt.plot(x, pdf, color='navy', linewidth=2)
    plt.title("Standard Normal Distribution (μ=0, σ=1)")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axvline(0, color='gray', linestyle='--')  # Mean line
    plt.show()

# Call the function
plot_standard_normal()
```

```python
#Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def compute_and_plot_zscore(x, mu, sigma):
    """
    Compute Z-score and plot its position on the standard normal distribution.

    Parameters:
    x : float       - Data point
    mu : float      - Mean of the distribution
    sigma : float   - Standard deviation
    """
    z = (x - mu) / sigma
    print(f"Z-score for x = {x} (μ = {mu}, σ = {sigma}): {z:.2f}")

    # Plot standard normal distribution
    x_vals = np.linspace(-4, 4, 500)
    y_vals = norm.pdf(x_vals, 0, 1)

    plt.plot(x_vals, y_vals, color='blue', label='Standard Normal PDF')
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='skyblue', alpha=0.5)
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z-value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
compute_and_plot_zscore(x=75, mu=70, sigma=10)
```

```python
#Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample, mu0, sigma, alpha=0.05):
    """
    Perform one-sample Z-test.

    Parameters:
    sample : array-like  - Sample data
    mu0 : float          - Population mean under null hypothesis
    sigma : float        - Known population standard deviation
    alpha : float        - Significance level (default 0.05)

    Returns:
    Z-score, p-value, and test conclusion
    """
    n = len(sample)
    x_bar = np.mean(sample)
    z = (x_bar - mu0) / (sigma / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))  # two-tailed p-value

    print(f"Sample mean = {x_bar:.2f}")
    print(f"Z-statistic = {z:.3f}")
    print(f"P-value = {p:.4f}")

    if p < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

    return z, p

# Example usage
sample_data = np.random.normal(loc=102, scale=15, size=50)
z_test(sample_data, mu0=100, sigma=15, alpha=0.05)
```

```python
#Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin

    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident that the true population mean lies within this interval.")

    return (lower, upper)

# Example usage
data = np.random.normal(loc=50, scale=10, size=30)
confidence_interval(data)
```

```python
#Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
from scipy.stats import t

# Step 1: Generate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=40)  # mean=100, std=15, n=40

# Step 2: Calculate the confidence interval
confidence = 0.95
n = len(data)
mean = np.mean(data)
std_err = np.std(data, ddof=1) / np.sqrt(n)
t_crit = t.ppf((1 + confidence) / 2, df=n-1)

margin = t_crit * std_err
lower, upper = mean - margin, mean + margin

# Output the result
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print("Interpretation: We are 95% confident the true population mean lies within this interval.")
```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt

# Sample datasets
data1 = np.random.normal(0, 1, 1000)  # mean=0, std=1
data2 = np.random.normal(0, 2, 1000)  # mean=0, std=2

# Calculate standard deviations
std1 = np.std(data1, ddof=1)  # ddof=1 for sample std deviation
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of data1: {std1:.3f}")
print(f"Standard Deviation of data2: {std2:.3f}")

# Visualization
labels = ['Dataset 1', 'Dataset 2']
std_devs = [std1, std2]

plt.bar(labels, std_devs, color=['blue', 'orange'])
plt.ylabel('Standard Deviation')
plt.title('Comparison of Standard Deviations')
plt.show()
```

```python
#Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample dataset
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate mean and std
mean = np.mean(data)
std = np.std(data)

# Z-score normalization
z_normalized = (data - mean) / std

# Plot original and normalized data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(z_normalized, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calc_skewness_kurtosis(size=1000, mean=0, std=1):
    # Generate normal distribution data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # By default, Fisher’s definition (excess kurtosis)

    return data_skewness, data_kurtosis

# Example usage
skew_val, kurt_val = calc_skewness_kurtosis()
print(f"Skewness: {skew_val:.4f}")
print(f"Kurtosis (excess): {kurt_val:.4f}")
```
